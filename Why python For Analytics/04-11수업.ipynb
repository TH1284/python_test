```python
selloutData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
selloutData.head(1)
```

```python
selloutData["WEEK"] = selloutData["YEARWEEK"]%100
selloutData.head()
```

```python
selloutData["YEAR"] = selloutData["YEARWEEK"].apply(customFunction)
selloutData
```

```python
selloutData["YEAR"] = round(selloutData["YEARWEEK"]/100)
selloutData
```

```python
refinedDate = selloutData[selloutData.YEARWEEK % 100 < 53]
refinedDate.head()
```

```python
selloutData["QTY_NEW"] = np.where(selloutData["QTY"]<0,0,selloutData["QTY"])
selloutData
```

```python
selloutData[selloutData.QTY >=0]
```

```python
sortKey = ["REGIONID", "PRODUCT", "YEARWEEK"]
```

```python
sortedDate = refinedDate.sort_values(sortKey)
sortedDate
```

```python
##인덱스 정리
step1Data = sortedDate.reset_index()
step1Data
```

```python
##인덱스 삭제 후 정리
step1Data = sortedDate.reset_index(drop = True)
step1Data
```

```python
##rolling함수 하용하기
step1Data.QTY_NEW.rolling(window=15, center=True).mean()
```

```python
step1Data["MA"] = step1Data.QTY_NEW.rolling(window=15, center=True).mean()
step1Data
```

```python
##min_periods를 쓰면 평균값을 모두 나타낼수있다(ex: window =5,min_periods=4가 되면 0번째 인뎃스 평균값이 나오지 안는다)
## 하지만 min_periods=1 or 2 or 3은 센터에서 포함이 되기 떄문에 평균값을 나타낼수 있다.
step1Data["MA"] = step1Data.QTY_NEW.rolling(window=5, min_periods = 1, center=True).mean()
step1Data
```

```python
##현재경로에 저장
step1Data.to_csv("./result.csv")
```

```python
groupKey = ["REGIONID", "PRODUCT", "YEAR"]
```

```python
groupData = step1Data.groupby(groupKey)[["QTY_NEW"]].mean()
groupData
```

```python
groupDataResult = groupData.reset_index()
```

```python
groupDataResult.head()
```

```python
##컬럼명 변경
groupDataResult = groupDataResult.rename(columns = {'QTY_NEW': 'QTY_MEAN'})
```

```python
groupDataResult.head()
```

```python
step1Data.head()
```

```python
##데이터 합치기
mergeResult = pd.merge(step1Data, groupDataResult,\
                      left_on = ["REGIONID","PRODUCT","YEAR"], right_on =["REGIONID","PRODUCT","YEAR"])
mergeResult.head()
```
