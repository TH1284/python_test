```python
def minmaxweek(inputdf) :
    minweek = min(inputdf['yearweek']) ## yearweek의 최솟값을 넣어준다
    maxweek = max(inputdf['yearweek']) ## yearweek의 최댓값을 넣어준다
    fullyearweek = [] ## 빈 배열을 만들어준다.
    i = minweek       ## minweek를 반복할 변수에 넣어준다
    fullyearweek.append(i) ## 빈 배열에 들어갈 데이터들을 append 해준다
    while (max(fullyearweek) < maxweek): ## 반복문으로 최댓값보다 최솟값이 작을 때 반복
        inputWeek = int(postWeek(str(max(fullyearweek)),1))
        fullyearweek.append(inputWeek)
    newyearweekList = list(set(fullyearweek).difference(set(list(inputdf.yearweek.values))))

    for i in range(len(newyearweekList)):
        inputdf.loc[len(inputdf)] = list(inputdf.loc[0].values)
        inputdf.loc[len(inputdf)-1,'yearweek'] = newyearweekList[i]
        inputdf.loc[len(inputdf)-1,'year'] = round(newyearweekList[i]/100)
        inputdf.loc[len(inputdf)-1,'week'] = newyearweekList[i]%100
        inputdf.loc[len(inputdf)-1,'qty'] = 0.0
    inputdf.sort_values('yearweek')
    inputdf.reset_index(drop = True, )
    return inputdf
```
