```python
import pandas as pd
```

```python
selloutData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
selloutData.head()
```

```python
selloutData.dtypes
```

```python
selloutData["WEEK"] = selloutData["YEARWEEK"]%100
selloutData.head()
```

```python
def Function(inputColumn):
    yearValue = inputColumn/100
    yearValue = int(yearValue)
    return yearValue
```

```python
selloutData["YEAR"] = selloutData["YEARWEEK"].apply(Function)
selloutData.head()
```

```python
refindData = selloutData[selloutData.YEARWEEK % 100 <53]
refindData.head()
```

```python
refindData["test"] = (selloutData.YEARWEEK/100).astype(int)
```

```python
refindData.head()
```

```python
import numpy as np
```

```python
selloutData["NEW_QTY"] = np.where(selloutData["QTY"]<0,0,selloutData["QTY"])
selloutData.head()
```

```python
selloutData[selloutData.QTY >=0]
```

```python
sortKey = ["REGIONID", "PRODUCT", "YEARWEEK"]
```

```python
sortedData = refindData.sort_values(sortKey)
sortedData.head()
```

```python
step1Data = sortedData.reset_index()
step1Data.head()
```

```python
step1Data = sortedData.reset_index(drop = True)
step1Data.head()
```

```python
step1Data["MA"] = step1Data.NEW_QTY.rolling(window = 15,min_periods = 1 , center = True).mean()
step1Data.head()
```

```python
groupKey = ["REGIONID", "PRODUCT", "YEAR"]
```

```python
groupData = step1Data.groupby(groupKey)[["NEW_QTY"]].mean()
groupData.head()
```

```python
groupDataResult = groupData.reset_index()
```

```python
groupDataResult.head()
```

```python
mergeResult = pd.merge(step1Data, groupDataResult,\
                      left_on = ["REGIONID","PRODUCT","YEAR"], right_on =["REGIONID","PRODUCT","YEAR"])
mergeResult.head()
```

```python
groupKey = ["REGIONID", "PRODUCT"]
```

```python
def groupRolling(oneGroup):
    indexGroupData = oneGroup.reset_index(drop = True)
    indexGroupData["MA2"] = indexGroupData["NEW_QTY_x"].rolling(window=15, center = True , min_periods =1).mean()
    return indexGroupData
```

```python
finalResult = mergeResult.groupby(groupKey).apply(groupRolling)
```
