```python
import pandas as pd
import numpy as np
```

```python
## 1.csv 파일을 불러온다
```

```python
actual_sales = pd.read_csv("./dataset/data/pro_actual_sales.csv")
actual_sales.head()
```

```python
## 2.필요한 컬럼만 추출한다
```

```python
actual_sales_cul = actual_sales[['regionSeg1','productSeg2','regionSeg2','productSeg3','yearweek','qty']]
actual_sales_cul.head()
```

```python
actual_sales_cul2 = actual_sales_cul
```

```python
groupKey = ['regionSeg1','productSeg2','regionSeg2','productSeg3','yearweek']
```

```python
groupData = actual_sales_cul2.groupby(groupKey)[["qty"]].mean()
groupData
```

```python
## 컬럼명 변경
```

```python
groupDataResult = groupData.rename(columns = {'qty': 'qty_mean'})
```

```python
groupDataResult
```

```python
groupDataResult = groupDataResult.reset_index()
```

```python
groupDataResult
```

```python
##데이터 합치기
```

```python
mergeResult = pd.merge(actual_sales_cul, groupDataResult,\
                      left_on = ["regionSeg1","productSeg2","regionSeg2","productSeg3","yearweek"], right_on =["regionSeg1","productSeg2","regionSeg2","productSeg3","yearweek"])
mergeResult.head()
```

```python
groupKey = ['regionSeg1','productSeg2','regionSeg2','productSeg3','yearweek']
```

```python
def groupRolling(oneGroup):

    #groupDataSet = mergeResult.groupby(groupKey)

    #oneGroup = groupDataSet.get_group(list(groupDataSet.groups)[2])

    indexGroupData = oneGroup.reset_index(drop = True)
    indexGroupData["MA"] = indexGroupData["qty"].rolling(window=15, center = True, min_periods=1).mean()

    return indexGroupData
```

```python
finalResult = mergeResult.groupby(groupKey).apply(groupRolling)
```

```python
finalResult
```

```python
finalResult2 = finalResult.reset_index(drop=True)
```

```python
finalResult2
```
