```python
import pandas as pd
import numpy as np
```

```python
## 데이터 불러오기
```

```python
actual_sales = pd.read_csv("./dataset/data/pro_actual_sales.csv")
actual_sales.head()
```

```python
## 필요한 컬럼명 추출
```

```python
actual_sales_cul = actual_sales[['regionSeg1','productSeg2','regionSeg2','productSeg3','yearweek','qty']]
actual_sales_cul.head()
```

```python
actual_sales_cul['qtymean'] = actual_sales_cul['qty'].ipynb_checkpoints/
```

```python
actual_sales_cul
```

```python
## 그룹비이
```

```python
groupKey = ['regionSeg1','productSeg2','regionSeg2','productSeg3','yearweek']
```

```python
groupData = actual_sales_cul.groupby(groupKey)[["qtymean"]].mean()
groupData
```

```python
groupData2 = actual_sales_cul.groupby(groupKey)[["qty"]]
groupData2
```

```python
## 컬럼명 변경 qty->  qty_mean
```

```python
groupData = groupData.rename(columns = {'qty': 'qty_mean'})
groupData
```

```python
## 평균 
```

```python
groupData["ma"] = groupData.qty_mean.rolling(window=15, center=True).mean()
groupData
```

```python
groupData.to_csv("./actual_groupby.csv")
```

```python
def groupRolling(oneGroup):

    #groupDataSet = mergeResult.groupby(groupKey)

    #oneGroup = groupDataSet.get_group(list(groupDataSet.groups)[2])

    indexGroupData = oneGroup.reset_index(drop = True)
    indexGroupData["MA2"] = indexGroupData["qty_mean"].rolling(window=15, center = True, min_periods=1).mean()

    return indexGroupData
```

```python
finalResult = groupData.groupby(groupKey).apply(groupRolling)
```

```python
finalResult
```

```python
finalResult.to_csv("./actual_groupby2.csv")
```
