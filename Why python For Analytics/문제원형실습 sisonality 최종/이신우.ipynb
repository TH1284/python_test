### step.2-1. 빈주차 생성

```python
## postWeek 함수 생성,(미래 주차 구하는 함수)

def postWeek(inputYW, gapWeek) :
    inputYW = str(inputYW)
    inputY = inputYW[0:4]
    inputW = inputYW[4:]
    
    gapWeek = int(gapWeek)
    
    sumWeek = int(inputW) + gapWeek
    
    currYear = int(inputY)
    lastWeek = Week.last_week_of_year(currYear).week  ## currYear, 마지막주차 구하는 함수
    
    while(sumWeek > lastWeek) :
        sumWeek -= lastWeek
        currYear += 1
        lastWeek = Week.last_week_of_year(currYear).week
        
    if(sumWeek < 10) :
        lastAnswer = str(currYear) + "0" + str(sumWeek)
    else :
        lastAnswer = str(currYear) + str(sumWeek)
    
    return lastAnswer
```

```python
## minmaxweek 함수 생성 : groupby를 적용한 데이터셋에 빈주차를 파악하여 최종적으로 채워주는 함수
## step1. yearweek의 min값과 max값을 찾아 그 사이에 있는 연주차를 모두 생성하여 배열에 담는다
## step2.비어있는주차가 존재하는 df와 step1에서 만든 배열의 차집합을 구하여 빈주차값을 알아본다
## step3. step2에서 구한 빈주차를 df에 적용시킨다(그룹별로 적용됨)

def minmaxweek(inputdf) : 
    inputdf.reset_index(drop = True, inplace = True)  ## 그룹바이 했을때 정리해줌,
    
    minYW = min(inputdf['yearweek']) 
    maxYW = 201627 ##전체 ITEM의 확정된 maxYW, 
    fullYearWeek = []  ## 빈 배열 or 리스트 만들어줌 체크해보자.
    fullYearWeek.append(minYW) ## 최소값들어감!
    
    while(max(fullYearWeek) < maxYW) :  ## fullYearWeek 의 max 값 까지  < 201627
        inputYW = int(postWeek(max(fullYearWeek),1))  ## 해당 리스트의 최대
        fullYearWeek.append(inputYW)
        
    diffYearWeek = list(set(fullYearWeek).difference(set(list(inputdf.yearweek.values)))) #difference는 set형식에서만 사용할 수 있음
    ## inputdf = 그룹바이 데이터 프레임의 yearweek 의 values 을 담아줌 차집합(difference)은 set 형식에서만사용가능하다.
    ## diffYearWeek 에는 없던 주차가 담겨있다.
    
    for i in range(len(diffYearWeek)) :  ## 비어있는 주차애들
        inputdf.loc[len(inputdf)] = list(inputdf.loc[0].values)  ## inputdf 데이터프레임의. loc[0] 제일 처음의 행의 값을 가져옴
        ## 위에 뽑은거를 예를들면, 데이타가 5개가 있어.. 길이는 5야 근데 loc 는 인덱스 번호를 가져온다. loc의 5라는 소린데,
        ## 실제로 인덱스 번호로 치면 0부터니까 4까지밖에 없죠, 새로 만들어줫다는 의미가 된다.
        ## 그룹단위로 들어오는 데이터를 
        
        inputdf.loc[len(inputdf)-1, 'yearweek'] = diffYearWeek[i]  
        inputdf.loc[len(inputdf)-1, 'year'] = int(diffYearWeek[i]/100) ##round 처리 말고 int처리로 변경
        inputdf.loc[len(inputdf)-1, 'week'] = diffYearWeek[i]%100
        inputdf.loc[len(inputdf)-1, 'qty'] = 0
    
    #어차피 이 함수 적용하고나서 다시 sort 해줘야하기때문에 생략해도 될 것 같음
#     inputdf.sort_values('yearweek')
#     inputdf.reset_index(drop = True, inplace = True)
    
    return inputdf
```

```python
# 다른 처리
# minmaxweek 함수 생성 : groupby를 적용한 데이터셋에 빈주차를 파악하여 최종적으로 채워주는 함수
# step1. yearweek의 min값과 max값을 찾아 그 사이에 있는 연주차를 모두 생성하여 배열에 담는다
# step2.비어있는주차가 존재하는 df와 step1에서 만든 배열의 차집합을 구하여 빈주차값을 알아본다
# step3. step2에서 구한 빈주차를 df에 적용시킨다(그룹별로 적용됨)

def minmaxweek(inputdf) :
    inputdf.reset_index(drop = True, inplace = True)  ## 
    
    minYW = min(inputdf['yearweek']) 
    maxYW = 201627 ##전체 ITEM의 확정된 maxYW 
    fullYearWeek = []
    fullYearWeek.append(minYW)
    
    while(max(fullYearWeek) < maxYW) :
        inputYW = int(postWeek(max(fullYearWeek),1))
        fullYearWeek.append(inputYW)
        
    diffYearWeek = list(set(fullYearWeek).difference(set(list(inputdf.yearweek.values)))) #difference는 set형식에서만 사용할 수 있음
    
    result2 = {     ## 빈 딕셔너리를 만들어준다. 왜?
        'regionSeg1': [],
        'productSeg1': [],
        'productSeg2': [],
        'regionSeg2': [],
        'regionSeg3': [],
        'productSeg3': [],
        'yearweek': [],
        'year': [],
        'week': [],
        'qty': []
    }
    OrderedDict(result2)  
    ## dict 형은 입력한 값 순서대로 출력되지 않고, 뒤죽박죽(알파벳 숫자, 뭐 그런식으로 ) -> OrderedDict() 사용하여 보정
    
    for i in range(len(diffYearWeek)) :  ## 빈주차만큼 돌린다.
        
#         result.append(list(inputdf.loc[0].values))
#         result['yearweek'].append(diffYearWeek[i])
#         result['year'].append(int(diffYearWeek[i]/100))
#         result['week'].append(diffYearWeek[i]%100)
#         result['qty'].append(0)
        
#         inputdf.loc[len(inputdf)] = list(inputdf.loc[0].values)
#         inputdf.loc[len(inputdf)-1, 'yearweek'] = diffYearWeek[i]
#         inputdf.loc[len(inputdf)-1, 'year'] = int(diffYearWeek[i]/100) ##round 처리 말고 int처리로 변경
#         inputdf.loc[len(inputdf)-1, 'week'] = diffYearWeek[i]%100
#         inputdf.loc[len(inputdf)-1, 'qty'] = 0
          
    
    
        result2['regionSeg1'].append(list(inputdf.loc[0].values)[0])  ## 이 부분의 loc[0] 때문에 처음에 reset_index 를 해줌
        result2['productSeg1'].append(list(inputdf.loc[0].values)[1])  ## 그룹단위의 데이터프레임으로 작업을 하는데 9만~12만건사이의
        result2['productSeg2'].append(list(inputdf.loc[0].values)[2])  ## 데이타프레임이 들어오면 인덱스 번호가 9만부터 시작되니까
        result2['regionSeg2'].append(list(inputdf.loc[0].values)[3])    ## 그런거임
        result2['regionSeg3'].append(list(inputdf.loc[0].values)[4])
        result2['productSeg3'].append(list(inputdf.loc[0].values)[5])  
        
        ## inputdf.loc[len(inputdf)] = list(inputdf.loc[0].values)
        ## 아까꺼(위에)를 풀어쓴건데 속도가 훨씬 빠름
        ## loc 이란걸 쓰면 느려진다.
        
        result2['yearweek'].append(int(diffYearWeek[i]))
        result2['year'].append(int(diffYearWeek[i]/100))
        result2['week'].append(int(diffYearWeek[i]%100))
        result2['qty'].append(0)
      
   
    result2=pd.DataFrame(result2)
    dfconcat=pd.concat([inputdf,result2]) ## 위아래로 붙이는거~
    dfconcat['qty'] = dfconcat['qty'].astype('int')
    dfconcat['yearweek'] = dfconcat['yearweek'].astype('int')
    dfconcat['year'] = dfconcat['year'].astype('int')
    dfconcat['week'] = dfconcat['week'].astype('int')    
    #어차피 이 함수 적용하고나서 다시 sort 해줘야하기때문에 생략해도 될 것 같음
#     inputdf.sort_values('yearweek')
#     inputdf.reset_index(drop = True, inplace = True)
    
    return dfconcat
```

## step3. 이동평균 구하기

```python
# 그룹단위별 rolling 함수 생성
def groupRolling(oneGroup) :
    indexGroupData = oneGroup.reset_index(drop = True)
    indexGroupData['MA'] = indexGroupData['qty'].rolling(window=4, center=False, min_periods=1).mean() 
    
    ##  window=4 가 날개를 펼친다, center=False 는 자기포함 과거 3주차, min_periods 페리우드, 난값 없애주는거
    ## 위에 값이 없으면 자기혼자만 포함해서 구함
    
    return indexGroupData
```

```python
groupKey2 = ['regionSeg1', 'productSeg2', 'regionSeg3']  ## 미국, 플로리다, 냉장고 
## 아이템별로 구할꺼기 때문에,
## 계절성지수 구할때는 별 차이는 없는데, 수요예측을 하게 되면, 새로운 아이템이라던지, 냉장고 종류가 추가되거나
## 없어질수가 있다, 세부 아이템까지 그룹키에 넣어버리면 없어져 버릴수도 있다. 
```

```python
maGroup = sortedData.groupby(groupKey2).apply(groupRolling) 
```

```python
def recent4Fcst (inputdf):
    
    inputdf.reset_index(drop = True, inplace = True)
    preweek1 = int(inputdf.loc[len(inputdf)-1].values[9])
    preweek2 = int(inputdf.loc[len(inputdf)-2].values[9])
    preweek3 = int(inputdf.loc[len(inputdf)-3].values[9])
    preweek4 = int(inputdf.loc[len(inputdf)-4].values[9])

    inputdf.loc[len(inputdf)] = list(inputdf.loc[0].values)

    # 식에 의해서 nan값이 나오게
    inputdf["fcst"] = (preweek1+preweek2+preweek3+preweek4)/4
    
    for i in range(len(inputdf))
        if (inputdf.loc[i] < inputdf.loc[(len(inputdf))]):
            inputdf["fcst"] = 0
    
    return inputdf
```
