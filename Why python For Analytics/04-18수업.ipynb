```python
import matplotlib.pyplot as plt
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 서브플롯
fig = plt.figure(figsize=(10,5))# 전체 크기
ax1 = fig.add_subplot(211)## 2행1열에 첫번째
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = plt.subplot(212)## 2행 1열에 두번째
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
#위는 판매량 아래는 계절성지수
```

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 이중축
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x,y, 'b--',lw=1.5, label = '1st')
ax1.plot(x,y, 'o')
ax1.grid(True)
ax1.legend(loc=2)
plt.ylabel('value')
plt.title('subplot')
ax2 = ax1.twinx()
ax2.plot(x,y2, 'g', lw=1.5, label='2nd')
ax2.plot(x,y2, 'go')
ax2.legend(loc=4)
plt.ylabel('promotion')
```

```python
sns.distplot(tips['total_bill'])
```

```python
## Jointplot 구현
```

```python
sns.jointplot(x= 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x= 'total_bill', y = 'tip', kind="kde",data = tips)
##kind="kde"추가시 커널 밀도함수 시연
```

```python
sns.pairplot(tips)

sns.pairplot(tips, hue = 'sex')


#hue라는 그룹바이를 이용함
#3차원 이상 데이터 시각화
```
