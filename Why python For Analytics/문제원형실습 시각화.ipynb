```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
featuresData = pd.read_csv("./dataset/ffffe.csv")
```

```python
featuresData
```

```python
refeaturesData = featuresData[featuresData.productSeg3 == 'ITEM0001']
refeaturesData
```

```python
refeaturesData
```

```python
refeaturesData2 = refeaturesData[refeaturesData.regionSeg3 == 'SITEID0001']
refeaturesData2
```

```python
refeaturesData3 = refeaturesData2[refeaturesData2.year == 2013]
refeaturesData3
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x=refeaturesData3[["week"]]
y=refeaturesData3[["qty"]]
y2=refeaturesData3[["MA"]]
y3=refeaturesData3[["seasonality"]]

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1)
plt.plot(x,y, 'b--',lw=1.5, label='1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('QTY')
plt.title('subplot')
ax2 = ax1.twinx()
ax2.plot(x,y2, 'g', lw=1.5, label='MA')
ax2.plot(x,y3, 'r', lw=1.5, label='seasonality')
# ax2.plot(x,y2, 'g')

ax2.legend(loc=1)
plt.ylabel('MA')
```

```python
import matplotlib.pyplot as plt
x=refeaturesData3[["week"]]
y =refeaturesData3[["qty"]]
y2 =refeaturesData3[["MA"]]
y3=refeaturesData3[["seasonality"]]

# 서브플롯
fig = plt.figure(figsize=(10,5))# 전체 크기
ax1 = fig.add_subplot(311)## 2행1열에 첫번째
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('qty')
plt.title('qty& MA & seasonality compare')
ax2 = plt.subplot(312)## 2행 1열에 두번째
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('MA')
ax3 = plt.subplot(313)## 2행 1열에 두번째
plt.plot(x,y3, 'r', lw=1.5, label='3nd')
plt.plot(x,y3, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('seasonality')
#위는 판매량 아래는 계절성지수
```
