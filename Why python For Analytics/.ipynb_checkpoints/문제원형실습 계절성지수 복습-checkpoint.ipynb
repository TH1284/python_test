```python
def minmaxweek(inputdf):
    inputdf.reset_index(drop = True, inplace = True)## 필요 없???
    minweek = min(inputdf['yearweek']) ## inputdf의 yearweek컬럼에 최솟값을 담은 부분
    maxweek = max(inputdf['yearweek']) ## inputdf의 yearweek컬럼에 최댓값을 담은 부분
    
    fullyearweek = [] ## 빈 배열을 만들어서 minweek부터 maxweek까지의 모든 yearweek를
                      ## 담아주려는 부분
    i = minweek ## 처음 시작부분을 최소 yearweek로 시작하기 위해서 나타낸 부분
    fullyearweek.append(i) ## 최소 주차를 추가하는 부분
    
    ## 모든 주차를 생성해서 fullyearweek에 담으려는 부분
    while (max(fullyearweek) < maxweek):
        inputWeek = int(postweek(str(max(fullyearweek)),1)) ##입력할 주차는 최대 주차 다음의 주차
        fullyearweek.append(inputWeek) ## fullyearweek에 담어주는 부분
    ## fullyearweek와 inputdf.yearweek.values의 차집합을 구하는 부분    
    newyearweekList = list(set(fullyearweek).difference(set(list(inputdf.yearweek.values))))
    
    ## newyearweekList의 길이 만큼 반복하여 inputdf에 비워진 주차를 추가하는 부분
    for i in range(len(newyearweekList)):
        inputdf.loc[len(inputdf)] = list(inputdf.loc[0].values) ## 첫번째 행을 가지고 복사함
                                                                ## 어느 행이든 상관없음
        inputdf.loc[len(inputdf)-1,'yearweek'] = newyearweekList[i] ## 컬럼yearweek의 i번째를 추가  
        inputdf.loc[len(inputdf)-1,'year'] = round(newyearweekList[i]/100) ## 컬럼year의 i번째를 추가  
        inputdf.loc[len(inputdf)-1,'week'] = newyearweekList[i]%100 ## 컬럼week의 i번째를 추가  
        inputdf.loc[len(inputdf)-1,'qty'] = 0 # 컬럼 값 변경
                           
    inputdf.sort_values('yearweek') ## yearweek컬럼을 정렬
    inputdf.reset_index(drop = True, inplace = True) ## index를 다시 해주는 부분
    return inputdf   
```
